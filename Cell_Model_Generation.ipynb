## Simulate batteries models with different SOC limitations in hourly resolution.

```python
soc_limitation_df = pd.DataFrame(columns=['Min', 'Max', 'n_soc_grid_points'], 
                                    data = [[0, 1, 51],
                                                [0.1, 0.9, 41],
                                                [0.2, 0.8, 31],
                                                [0.3, 0.7, 21],
                                                [0.4, 0.6, 11]])
```

```python
soc_limitation_df
```

```python
def create_cells_hourly(df):
    soc_min = df['Min']
    soc_max = df['Max']
    n_soc_grid_points = df['n_soc_grid_points']

    cache_dir = f"./battery_model_params_20260214/1_hour_{soc_min}_{soc_max}"


    cell_hourly = CellSLB(
        file_path="./SLB_para.xlsx",
        dt_step_s=3600.0,
        soc_init=0.5,
        soc_min=soc_min,
        soc_max=soc_max,
        max_charge_current=0.84,
        max_discharge_current=-0.84,
        use_average_power_method=True,
        n_power_segments=120,
        max_opt_iter=50,
        opt_tol=0.005,
        use_parallel_computation=True,
        n_jobs=-1,
        cache_dir=cache_dir,  # 设置缓存目录
        use_cache=True,  # 启用缓存
        n_soc_grid_points=n_soc_grid_points,
    )

    # 手动保存power limit（可选，如果参数一致会自动保存）
    cell_hourly.save_power_limits()
    return cache_dir
```

```python
r = soc_limitation_df.apply(create_cells_hourly, axis=1)
```
